```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
show_num = 10
train_raw_dir = './dataset/binarized_mnist_train.amat'
train_img_dir = './dataset/train/'
```

```python
train_raw_data = np.loadtxt(train_raw_dir)
train_imgs = sorted(os.listdir(train_img_dir))
```

```python
plt.figure(figsize=(18, 7))
plt.rcParams['axes.titlesize'] = 20
plt.axis('off')

for i in range(show_num):
    temp = train_raw_data[i].reshape((28, 28))
    plt.subplot(2, 5, i+1)
    img = plt.imshow(temp); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
```

```python
plt.figure(figsize=(18, 7))
plt.rcParams['axes.titlesize'] = 20
plt.axis('off')

for i in range(show_num):
    temp = mpimg.imread(train_img_dir + train_imgs[i])
    plt.subplot(2, 5, i+1)
    img = plt.imshow(temp); img.axes.get_xaxis().set_visible(False); img.axes.get_yaxis().set_visible(False)
```
